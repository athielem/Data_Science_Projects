The following datasets contain **information on cycling activities in Munich** based on district; a *Raddauerzählstelle* is a roughly 2 meters high metallic device that is spread throughout Munich and continuously records the number of passing cyclists (including the direction from where the cyclist is coming from).

The datasets can be found under the following **link**: https://opendata.muenchen.de/dataset/daten-der-raddauerzaehlstellen-muenchen-2022. This is the official *repository* of the city of Munich with a variety of different datasets that have been recorded over time. It seems plausible to assume that the underlying reasoning for making this data public lies in satisfying growing demand from both private and public initiatives to perform data mining with information that was previously not available or technically attainable (the beforementioned recording devices can be viewed as progress on local digitalization measures) - for the purpose of gaining new insights.

Let's take a look at the **two datasets**:

```python
# Creating dataframes with CSV files
import pandas as pd

csv_url_minutes = "https://raw.githubusercontent.com/athielem/Data_Science_Projects/main/Datasets/rad_2022_01_15min.csv"
csv_url_days = "https://raw.githubusercontent.com/athielem/Data_Science_Projects/main/Datasets/rad_2022_01_tage.csv"

df_minutes = pd.read_csv(csv_url_minutes)
df_days = pd.read_csv(csv_url_days)
```
